```python
import pandas as pd
#df_pr= pd.read_csv('./Stock Data/indexProcessed.csv')
df= pd.read_csv('./Stock Data/indexData.csv')
dfinfo= pd.read_csv('./Stock Data/indexInfo.csv')
#print(df.columns)
#print(dfinfo.columns)
df_temp= pd.merge(df,dfinfo)
dfind= df_temp.loc[df_temp['Region']=='India']
col= list(dfind.columns)
dfind=dfind[col[:2]+[col[9]]+col[2:8]]
dfind
```

```python
dfny= df_temp.loc[df_temp['Index']=='NYA']
col= list(dfny.columns)
dfny=dfny[col[:2]+[col[9]]+col[2:8]]
dfny
```
