# Question 1:

```python
df= pd.read_csv('test1.csv')

df.drop(['Id'], inplace=True, axis=1)
```

```python
lis= list((df.isna().sum()/df.shape[0])*100)
print("the dropped columns: ")
for x in lis:
    if(x>30):
        print(df.columns[lis.index(x)])
        df.drop(df.columns[lis.index(x)], inplace=True, axis=1)
        
```

```python
sns.distplot(df["SalePrice"])
plt.show();
```

```python
plt.rcParams['figure.figsize']= (10,10)
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
df2= df.select_dtypes(include= numerics)
df2.hist()
plt.show();
```

```python
x= df2.corr()
x.sort_values(by= 'SalePrice', ascending=False, inplace=True)
print("The following 10 strongly correlated columns to SalePrice except SalePrice itself")
print(x['SalePrice'].iloc[1:11])
```

```python
%matplotlib inline
g = sns.PairGrid(df2, y_vars= ['SalePrice'])
g.map(sns.scatterplot)
plt.show();
```
