```python
df.drop(columns= ["Unnamed: 32"],inplace=True)
```

```python
X= df.drop(columns= ['diagnosis'])
y= df['diagnosis']
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X= pd.DataFrame(scaler.fit_transform(X), columns= X.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
labe= LabelEncoder()
y= labe.fit_transform(y)
```

```python
from sklearn.decomposition import PCA
pca_model= PCA(n_components= 2)
trans= pca_model.fit_transform(X)
```

```python
print(pca_model.explained_variance_ratio_)
```

### Graph showing transformed data against Dep. Variable

```python
plt.style.use('dark_background')
plotter= plt.scatter(trans[:, 0], trans[:, 1], c= y)
plt.legend(handles=plotter.legend_elements()[0], 
           labels=list(["B", "M"]))
plt.show();
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

### Apply Log. Regression to transformed data

```python
from sklearn.model_selection import train_test_split
Xtr, Xt, Ytr, Yt= train_test_split(trans, y, test_size= 0.2, stratify= y)
log_model= LogisticRegression()
log_model.fit(Xtr, Ytr)
preds= log_model.predict(Xt)
print("Accuracy: ",accuracy_score(preds, Yt))
print("Recall_Score: ",recall_score(preds, Yt))
print("Precision_Score: ",precision_score(preds, Yt))
```
