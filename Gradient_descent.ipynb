## finding the perfect degree for the GLM:


```python
n_orders= 6
for k in range(1, n_orders+1):
    fit_m= np.polyfit(dfdius, dfmpgus, k)
    vals= np.polyval(fit_m, dfdius)
    sse= np.sum((vals- dfmpgus)**2)
    bayes_k= m*np.log(sse)+ k*np.log(m)
    plt.plot(k, bayes_k, 'ro')
plt.show();
```

```python
desmat_quad= np.vstack((np.ones(m), dfdius, dfdius**2)).T
beta= np.linalg.lstsq(desmat_quad, dfmpgus, rcond= None)[0]
vals= desmat_quad@beta
```
